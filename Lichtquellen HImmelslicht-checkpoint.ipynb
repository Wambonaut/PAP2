```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def comma_to_float(valstr):
    return float(valstr.replace(',','.'))
```

```python
lamb_og, inten_og=np.loadtxt(r'Himmel auf.txt', skiprows=17, encoding="latin-1",
                             converters= {0:comma_to_float, 1:comma_to_float},
                            comments='>', unpack=True)
lamb_mg, inten_mg=np.loadtxt(r'himmel zu.txt', skiprows=17, encoding="latin-1",
                             converters= {0:comma_to_float, 1:comma_to_float},
                            comments='>', unpack=True)
lamb_s, inten_s=np.loadtxt(r'Sonne.txt', skiprows=17, encoding="latin-1",
                             converters= {0:comma_to_float, 1:comma_to_float},
                            comments='>', unpack=True)
```

```python

plt.plot(lamb_og, inten_og, label='ohne Fenster')
plt.plot(lamb_mg, inten_mg, label='mit Fenster')
plt.plot(lamb_s, inten_s, label='Sonne')
plt.title('Gemessenes Sonnespektrum direkt in Sonne, sowie mit und ohne Fenster')
plt.xlabel('Wellenlnge \nm')
plt.ylabel('Intensitaet \ b.E.')
plt.legend()
plt.grid()
plt.ylim((0,60000))
plt.xlim((250,900))
plt.show()
```

```python
##Bestimmen der Fraunhofer Linien
from scipy import signal
(minima,)=signal.argrelmin(inten_s,order=39)
minima=minima[10:-2]
plt.title("Fraunhofer Linien")
plt.xlabel("Wellenlnge")
plt.ylabel("Intensitt")
plt.plot(lamb_s, inten_s, label='Sonne', color="red")
plt.scatter(lamb_s[minima], inten_s[minima], label="Wellenlngen Linien")
plt.xlim(350,850)
plt.legend()
for i in minima:
    plt.annotate(str(lamb_s[i]),(lamb_s[i],inten_s[i]+1000))
```

```python
A=1-inten_mg/inten_og
plt.plot(lamb_mg,A)
plt.title('Absorption von Glas')
plt.xlabel('Wellenlaenge / nm')
plt.ylabel('Absorption / b.E.')
plt.ylim((0,1))
plt.xlim((320,800))
plt.show()
```

```python

plt.plot(lamb_og, inten_og, label='ohne Fenster')
plt.title('Spektrum ohne Sonne')
plt.xlabel('Wellenlnge \nm')
plt.ylabel('Intensitaet \ b.E.')
plt.xlim((350,800))
plt.figure(figsize=(30, 30), dpi= 80, facecolor='w', edgecolor='k')
plt.show()
```

```python

plt.plot(lamb_og,inten_og, label='ohne Fenster')
plt.plot(lamb_s,inten_s,label='Sonnenlicht')
plt.title('Sonnenlicht und Streulicht')
plt.xlabel('Wellenlaenge / nm')
plt.ylabel('Intensitt / b.E.')
plt.legend()
plt.xlim((320,800))
plt.show()
```

```python

B=inten_og/(inten_s+1)
plt.plot(lamb_mg,B)
plt.title('Verschiebung Himmel durch Sonnenlicht ')
plt.xlabel('Wellenlaenge / nm')
plt.ylabel('Absorption / b.E.')
plt.xlim((320,800))
plt.ylim((0,10))
plt.show()
```
